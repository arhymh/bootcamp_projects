```python
# profit by segment
print(df.groupby('Segment')['Profit']\
    .agg(['min','mean','max','sum']).sort_values('sum', ascending = False))
print("---------------------------------------------")
print(f"Result; 'Consumer' segment has the highest profit")
```

```python
# profit by category
print(df.groupby('Category')['Profit']\
    .agg(['min','mean','max','sum']).sort_values('sum', ascending = False))
print("---------------------------------------------")
print(f"Result; 'Technology' category has the highest profit")
```

```python
# Which product has the highest/lowest profit in Consumer-technology segment
techcon = df.query("Segment == 'Consumer' & Category == 'Technology'")
print(techcon.groupby('Sub-Category')['Profit']\
    .agg(['min','mean','max','sum']).sort_values('sum', ascending = False))
print("---------------------------------------------")
print(f"Result; In Consumer-technology, which is the most profitable segment,\
                \nthe highest profitable sub-category is 'Copiers'\
                \nthe lowest profitable sub-category is 'Machines'")
```

```python
# TODO 08 - what is the proportion of total sales (%) in West + Central in 2019 e.g. 25%
nt = df[pd.DatetimeIndex(df['Order Date']).year == 2019]
total = nt['Sales'].sum().round(2)
wctotal = nt.query("Region == 'West' | Region == 'Central'")['Sales'].sum().round(2)
percent = (wctotal/total)*100

print(f"Total Sale: {total}")
print(f"West+Central Sale: {wctotal}")
print("---------------------------------------------")
print(f"The proportion of total sale in West+Central region in 2019:\n{percent.round(2)} %")
```

```python
# TODO 09 - find top 10 popular products in terms of number of orders vs. total sales during 2019-2020
nttw = df[(df['Order Date'].dt.year >= 2019) & (df['Order Date'].dt.year <= 2020)]

#top 10 by sale
bysale = nttw.groupby('Product Name')['Sales']\
    .agg(['sum','count']).sort_values('sum', ascending = False)\
    .round(2).head(10).reset_index()
bysale.columns = ['Top 10 Products by sales', 'sum of sales', 'number of orders']

#top 10 by order
byorder = nttw.groupby('Product Name')['Sales']\
    .agg(['count','sum']).sort_values('count', ascending = False)\
    .round(2).head(10).reset_index()
byorder.columns = ['Top 10 Products by number of orders', 'number of orders', 'sum of sales']

# concat 2 list
byorder_vs_bysale = pd.concat([byorder,bysale], axis = 1)
byorder_vs_bysale
```

```python
# TODO 10 - plot at least 2 plots, any plot you think interesting :)
# plot 1: Sales trend for each region by year
# plot 2: What drive the profit in West region in 2019-2020?; Office supplies and Technology
```

```python
# plot 1: Sales trend for each region by year

# create year column from order date
df['Year'] = df['Order Date'].dt.strftime('%Y')
source = df

# grouping data for plot
plot = source.groupby(['Year','Region'])['Profit'].sum().round(2).reset_index()
print(source.groupby(['Year','Region'])['Profit'].sum().round(2))

# plot graph
plot.pivot(columns = 'Region', index = 'Year', values = 'Profit')\
    .plot(kind = 'line', xlabel = 'Year', ylabel = 'Profit amount',\
          title = 'Profit in each region from 2017-2020',figsize=(9,6));
```

```python
# plot 2: What drive the profit in West region in 2019-2020?

# change year to int data type
source['Year'] = source['Year'].apply(pd.to_numeric)

# query data
# 2019
westn = source.query("Region == 'West' & Year == 2019")
nt = westn.groupby(['Category','Year'])['Profit']\
    .agg(['sum']).round(2).sort_values('sum', ascending = False).reset_index()
# 2020
westtw = source.query("Region == 'West' & Year == 2020")
tw = westtw.groupby(['Category','Year'])['Profit']\
    .agg(['sum']).round(2).sort_values('sum', ascending = False).reset_index()

# concat 2 list
westnttw = pd.concat([nt,tw], axis = 0)
westnttw.columns = ['Category', 'Year', 'Profit']

# plot graph
westnttw.pivot(columns='Year', index='Category', values='Profit')\
    .plot(kind='bar', figsize=(9,6), ylabel='Total Profit', xlabel='Category',\
    title='West region profit by category in 2019-2020');

print(f"Office 'Supplies' and 'Technology' drive profit in West region")
westnttw
```
